## Load the Model and Tokenizer

```python
from unsloth import is_bfloat16_supported, FastLanguageModel

model_name = "unsloth/DeepSeek-R1-Distill-Qwen-1.5B-unsloth-bnb-4bit"
max_seq_length = 512

# Load the 4bit pre quantized model of deepseek and the tokenizer
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = model_name,
    max_seq_length = max_seq_length,
    dtype = None,
    load_in_4bit = True,
)
```

## Run the model

```python
prompt_style = """Below is an instruction and a question for context. Write a logical and accurate response.

### Instruction:
Detect the question's language and answer in the same language. Think step-by-step to ensure clarity and correctness.

### Question:
{}

### Response:
{}
"""
```

```python
# Run inference
question = "Is 1.11 bigger 1.2?"

FastLanguageModel.for_inference(model)
inputs = tokenizer([prompt_style.format(question, "")], return_tensors="pt").to("cuda")


outputs = model.generate(
    input_ids=inputs.input_ids,
    attention_mask=inputs.attention_mask,
    max_new_tokens=2048,
    use_cache=True,
)
response = tokenizer.batch_decode(outputs)
```
